### Part (a) Convolutional Network - 3 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

**Ans**

I decided to build a (convolution network) CNN and (fully connected Network) FCN. For CNN I have selected to have 3 convolution layers, first layer with 8 output channels, kernel size of 3X3 and stride/padding of 1. For second layer, I have 16 output channels, kernel size of 3X3 and stride/padding of 1, For third layer, I have 32 output channels, kernel size of 1X1 and stride/padding of 1. My understanding is we are to increase the number of channels to increase number of filters. Each channel is looking at different expects of the images. The kernel size is meant to pick up different feature of the image and having small kernel size will help to focus in one smaller area and it will not miss any small details. Last layer of 1X1 is meant to sperate all the important features and little differences between pixel/image features. 

The FCN only has 2 layers, input layers take in the features selected by CNN and classifies them into 9 classes. I have selected to have only 50 hidden units because this will help concisely combines all the feature within 50 layers which are then split into 9 classes.


```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from statistics import mean
```

```python
class handgesture(nn.Module):
    
    
    # Defining the Constructor
    def __init__(self, n_classes=9):
        super(handgesture, self).__init__()
        
        # First convolutional layer takes 3 inputs and generates 24 outputs
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3, stride=1, padding=1)
        
        # Second convolutional layer takes 24 inputs and generates 32 outputs
        self.conv2 = nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, stride=1, padding=1)
        
        # Third convolutional layer takes 32 input channels, and generates 64 outputs
        self.conv3 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=1, stride=1, padding=1)

        # max pooling with a kernel size of 2
        self.pool = nn.MaxPool2d(kernel_size=2)
        
        # drop layer deletes 20% of the features to help prevent overfitting
        #self.drop = nn.Dropout2d(p=0.2)
        
        #fc layers 1
        self.fc1 = nn.Linear(in_features=29 * 29 * 32, out_features=50)

        #fc layers 2
        #self.fc2 = nn.Linear(in_features=500, out_features=200)

        #self.fc3 = nn.Linear(in_features=200, out_features=100)

        self.fc4 = nn.Linear(in_features=50, out_features=n_classes)

    def forward(self, x):
        # ReLU activation function layer 1
        x = self.pool(F.relu(self.conv1(x))) 
        # ReLU activation function layer 2
        x = self.pool(F.relu(self.conv2(x))) 
        # ReLU activation function layer 3
        x = self.pool(F.relu(self.conv3(x)))  
        # dropout to prevent overfitting 
        #x = F.dropout(self.drop(x), training=self.training)
        # Flatten
        x = x.view(-1, 29 * 29 * 32)
        # fully-connected layers
        x = F.relu(self.fc1(x))
        #x = F.relu(self.fc2(x))
        #x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x


# Create an instance of the model class and allocate it to the device
model = handgesture(n_classes=9)

print(model)
```

### Part (b) Training Code - 4 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

**Ans**

Below is the model accuracy and train functions, please note this function were reused from the **Tutorial 2B Transfer learning**. Most of the function is very similar, but some parts are modified to suit and meet the requirements. Changes such as easy tuning of the hyperparameters and model check points with print statements. In addition, Loss function was selected to be cross entropy loss and optimizer was selected to be SGD. Loss function was cross entropy because this is a multiclass classification model and cross entropy loss function will help the model to calculate the loss for all the classes. The optimizer was selected as SGD is better at generalization than Adam. For this lab generalization is important therefore SGD will be used.

```python
def get_accuracy(model, train_loader, val_loader, train=False):
           
    if train:
        data_loader = train_loader
    else:
        data_loader = val_loader

    correct = 0
    total = 0
    for imgs, labels in data_loader:

        if torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
           
        output = model(imgs)
        
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    return correct / total
```

```python
def train(model, train_set, val_set, batch_size=1, num_epochs=1, lr=0.01):
    
    train_loader = torch.utils.data.DataLoader(dataset=train_set, batch_size=batch_size, 
                                               shuffle=True)
    val_loader = torch.utils.data.DataLoader(dataset=val_set, batch_size=batch_size,
                                             shuffle=True)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)

    iters, losses, train_acc, val_acc = [], [], [], []

    # training
    n = 0 # the number of iterations
    torch.cuda.empty_cache()
    for epoch in range(num_epochs):
        #torch.cuda.empty_cache()
        for imgs, labels in iter(train_loader):
          
            #To Enable GPU Usage
            if torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################
            
              
            out = model(imgs)
            #print(out)             # forward pass
            #pred = out.max(1, keepdim=True)[1]
            loss = criterion(out, labels)
            #print(labels) # compute the total loss
            optimizer.zero_grad()         # a clean up step for PyTorch
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            
            
            # save the current training information
            iters.append(n)
            losses.append(float(loss)/batch_size)             # compute *average* loss
            train_acc.append(get_accuracy(model, train_loader, val_loader, train=True)) # compute training accuracy 
            val_acc.append(get_accuracy(model, train_loader, val_loader, train=False))  # compute validation accuracy
            n += 1
        print('Epoch: ', epoch, 'and Iteration ', n, 'Average Train Accuracy ', train_acc[-1], 'Average Val Accuracy ', val_acc[-1], 'Average Train Loss', losses[-1])
          
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
```

Train your new network, including any hyperparameter tuning. Plot and submit the training curve of your

Note: Depending on how you are caching (saving) your AlexNet features, PyTorch might still be tracking
updates to the **AlexNet weights**, which we are not tuning. One workaround is to convert your AlexNet
feature tensor into a numpy array, and then back into a PyTorch tensor.

```python
def get_accuracy(model, features_train, features_val, train=False):
    if train:
        feature = features_train
    else:
        feature = features_val

    correct = 0
    total = 0
    correct_list = []
    labels_list = []
    for imgs, labels in feature:
        
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
        #############################################
              
        output = model(imgs)
        
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
        accuracy = correct / total
    return accuracy
```

```python
def train(model, train_set, val_set, batch_size=20, lr= 0.01, num_epochs=1):
    
    train_loader = torch.utils.data.DataLoader(dataset=train_set, batch_size=batch_size, 
                                               shuffle=True)
    val_loader = torch.utils.data.DataLoader(dataset=val_set, batch_size=batch_size,
                                             shuffle=True)
    features_train = []
    features_val = []
    for p, i in enumerate(train_loader, 0):
      img, labels = i
      img = alexnet.features(img)
      feature_stack = torch.stack([f for f in img])
      labels_stack = torch.stack([l for l in labels])
      features_train.append((feature_stack, labels_stack))

    for p, i in enumerate(val_loader, 0):
      img, labels = i
      img = alexnet.features(img)
      feature_stack = torch.stack([f for f in img])
      labels_stack = torch.stack([l for l in labels])
      features_val.append((feature_stack, labels_stack))
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9)

    iters, losses, train_acc, val_acc = [], [], [], []
    
    # training
    n = 0 # the number of iterations
    for epoch in range(num_epochs):
        for imgs, labels in iter(features_train):
            
            imgs = torch.from_numpy(imgs.detach().numpy())  
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################
            
              
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch

            # save the current training information
            iters.append(n)
            losses.append(float(loss)/batch_size)             # compute *average* loss
            train_acc.append(get_accuracy(model, features_train, features_val, train=True)) # compute training accuracy 
            val_acc.append(get_accuracy(model, features_train, features_val, train=False))  # compute validation accuracy
            n += 1
        print('Epoch: ', epoch, 'and Iteration ', n, 'Average Train Accuracy ', train_acc[-1], 'Average Val Accuracy ', val_acc[-1], 'Average Train Loss', losses[-1])
    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
```
